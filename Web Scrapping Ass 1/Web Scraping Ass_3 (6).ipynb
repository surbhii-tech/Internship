```python
url=[]

start=0
end=3
for page in range(start,end):
    url_1=driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for i in url_1:
        a=i.get_attribute('href')
        url.append(a) 
        time.sleep(1)
    nxt=driver.find_element(By.XPATH,"//a[text()='Next']")
    nxt.click()
    time.sleep(1)
len(url)
```

```python
Brand=[]
Name=[]
Price=[]
Return_Exchange=[]
Exp_delivery=[]
Availability=[]
```

```python
for i in url:
    driver.get(i)
    time.sleep(2)
    
    try:
        brand = driver.find_element(By.XPATH,'//tr[@class="a-spacing-small po-brand"]')                           #Extracting Brand
        Brand.append(brand.text)

    except NoSuchElementException:
        Brand.append('-')
        
    time.sleep(1)

# Extracting name from id

    try:
        name = driver.find_element(By.XPATH,'//span[@id="productTitle"]')      
        Name.append(name.text)

    except NoSuchElementException:
        Name.append('-')
        
    time.sleep(1)

# Extracting Price

    try:
        price = driver.find_element(By.XPATH,'//span[@class="a-price-whole"]')      
        Price.append(price.text)

    except NoSuchElementException:
        Price.append('-')
    
    time.sleep(1)

# Extracting Return_Exchange

    try:
        repalcement = driver.find_element(By.XPATH,'//div[@class="a-column a-span3 a-text-center icon-container icon-grid-width celwidget"]')      
        Return_Exchange.append(repalcement.text)

    except NoSuchElementException:
        Return_Exchange.append('-')
        
    time.sleep(1)

# Extracting Exp_Delivery

    try:
        delivery= driver.find_element(By.XPATH,'//span[@class="a-text-bold"]')      
        Exp_delivery.append(delivery.text)

    except NoSuchElementException:
        Exp_delivery.append('-')

    time.sleep(1)

# Extracting Availability

    try:
        available= driver.find_element(By.XPATH,'//span[@class="a-size-medium a-color-success"]')      
        Availability.append(available.text)

    except NoSuchElementException:
        Availability.append('-')
```

```python
print(len(Brand),len(Name),len(Brand),len(Price),len(Brand),len(Return_Exchange),len(Exp_delivery),len(Availability),len(url))
```

```python
df=pd.DataFrame({'Brand Name':Brand,'Name of Product':Name,'Price':Price,'Return/Exchange':Return_Exchange,'Expected Delivery':Exp_delivery,'Availability':Availability,'Product URL':url})
df
```

```python
for i in prod_url:
    driver.get(i)
    time.sleep(1)
    
#Extracting Names
    try:
        name= driver.find_element(By.XPATH,'//span[@class="B_NuCI"]') 
        Name.append(name.text.split('(')[0])

    except NoSuchElementException:
        Name.append('-')
        
#Extracting Color
    try:
        col= driver.find_element(By.XPATH,'//span[@class="B_NuCI"]')
        col1=(col.text.split('(')[1])
        Color.append(col1.split(',')[0])

    except NoSuchElementException:
        Color.append('-')

        
#Extracting RAM  
    try:
        ram= driver.find_element(By.XPATH,'//div[@class="_2cM9lP"]')
        ram1=(ram.text.split('\n')[1])
        RAM.append(ram1.split('|')[0])

    except NoSuchElementException:
        RAM.append('-')

#Extracting ROM  
    try:
        rom= driver.find_element(By.XPATH,'//div[@class="_2cM9lP"]')
        rom1=(rom.text.split('\n')[1])
        ROM.append(rom1.split('|')[1])
    except NoSuchElementException:
        ROM.append('-')

#Extracting Camera
    try:
        cam= driver.find_element(By.XPATH,'//div[@class="_2cM9lP"]')
    
        Rear_Camera.append(cam.text.split('\n')[3])
    except NoSuchElementException:
        Rear_Camera.append('-')


#Extracting Display
    try:
        dis= driver.find_element(By.XPATH,'//div[@class="_2cM9lP"]')
        Display.append(dis.text.split('\n')[2])
    except NoSuchElementException:
        Display.append('-')

#Extracting Battrey
    try:
        battery= driver.find_element(By.XPATH,'//div[@class="_2cM9lP"]')
        Battery.append(battery.text.split('\n')[4])
    except NoSuchElementException:
        Battery.append('-')

#Extracting Price
    try:
        price= driver.find_element(By.XPATH,'//div[@class="_25b18c"]')
        Price.append(price.text)
    except NoSuchElementException:
        Price.append('-')
```

```python
url=[]
HOSTEL=[]
DISTANCE=[]
Ratings=[]
Total_Reviews=[]
Overall_Reviews=[]
Privates=[]
Dorms=[]
Facilities=[]
Description=[]
```

```python
#Extracting distance from citycenter
try:
    citycenter=driver.find_elements(By.XPATH,'//span[@class="description"]')
    for i in citycenter:
        DISTANCE.append(i.text.split('-')[1])
except NoSuchElementException:
    DISTANCE.append(citycenter.text)
```

```python
#Extracting Facilities
try:
    facility=driver.find_elements(By.XPATH,'//div[@class="facilities-label facilities"]')
    for i in facility:
        Facilities.append(i.text.split('\n')[0:])
except NoSuchElementException:
    Facilities.append(citycenter.text)
```

```python
url_1=driver.find_elements(By.XPATH,'//h2[@class="title title-6"]//a')
for i in url_1:
    url.append(i.get_attribute('href'))
```

```python
for i in url:
    driver.get(i)
    time.sleep(1)
    
    
#Extract Names
    try:
        name=driver.find_element(By.XPATH,'//div[@class="title-2"]')
        HOSTEL.append(name.text)
    except NoSuchElementException:
        HOSTEL.append('-')
        
#Extracting Rating
    try:
        rating=driver.find_element(By.XPATH,'//div[@class="score orange big"]')
        Ratings.append(rating.text)
    except NoSuchElementException:
        Ratings.append('-')

#Extracting Reviews
    try:
        Reviews=driver.find_element(By.XPATH,'//div[@class="reviews"]')
        Total_Reviews.append(Reviews.text)
    except NoSuchElementException:
        Total_Reviews.append('-')

#Extracting Overall Reviews
    try:
        Reviews=driver.find_element(By.XPATH,'//div[@class="keyword"]')
        Overall_Reviews.append(Reviews.text)
    except NoSuchElementException:
        Overall_Reviews.append('-')

#Extracting Description
    try:
        des=driver.find_element(By.XPATH,'//div[@class="content collapse-content"]')
        Description.append(des.text)
    except NoSuchElementException:
        Description.append('-')
```

```python
print(len(HOSTEL),len(DISTANCE),len(Ratings),len(Total_Reviews),len(Overall_Reviews),len(Facilities),len(Description),len(url))
```
