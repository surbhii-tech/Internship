```python
driver=webdriver.Chrome(r"F:\Internship\chromedriver_win32\chromedriver.exe")
```

```python
driver.get("https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos")
```

```python
for _ in range(10):
    driver.execute_script("window.scrollBy(0,100)")
    time.sleep(1)
```

```python
driver.maximize_window()
```

```python
Rank=[]
Name=[]
Artist=[]
Upload_Date=[]
Views=[]
```

```python
#Extracting Rank:

try:
    rank=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[1]')
    for i in rank[0:30]:
        Rank.append(i.text)
except NoSuchElementException:
    Rank.append('-')
    

#Extracting Video Name:

try:
    name=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[2]')
    for i in name[0:30]:
        Name.append(i.text.split('[')[0])
except NoSuchElementException:
    Name.append('-')
    
    
#Extracting Artist:

try:
    artist=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[3]')
    for i in artist[0:30]:
        Artist.append(i.text)
except NoSuchElementException:
    Artist.append('-')


    
#Extracting Upload Date:   
try:
    date=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[5]')
    for i in date[0:30]:
        Upload_Date.append(i.text)
except NoSuchElementException:
    Upload_Date.append('-')
    
    
#Extracting Views:   
try:
    view=driver.find_elements(By.XPATH,'//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[4]')
    for i in view[0:30]:
        Views.append(i.text)
except NoSuchElementException:
    Views.append('-')
```

```python
print(len(Rank),len(Name),len(Artist),len( Upload_Date),len(Views))
```

```python
df=pd.DataFrame({'Video Rank':Rank,'Video Name':Name,'Artist':Artist,'Upload Date':Upload_Date,'Views(Billions)':Views})
df
```

```python
#Extracting all Links

url=[]

link=driver.find_elements(By.XPATH,'//h1[@class="h3 lh-condensed"]/a')
for i in link:
    url.append(i.get_attribute('href'))
```

```python
Repository=[]
Description=[]
Count=[]
Language=[]
```

```python
for i in url:
    driver.get(i)
    time.sleep(2)
    
#Scraping Title-

    try:
        title = driver.find_element(By.XPATH,'//div[@class="flex-auto min-width-0 width-fit mr-3"]')
        Repository.append(title.text.split('\n')[0:3])

    except NoSuchElementException:
        Repository.append('-') 
   
        
#Scraping Description -


    try: 
        des = driver.find_element(By.XPATH,'//p[@class="f4 my-3"]')
        Description.append(des.text)
    except NoSuchElementException:
        Description.append('-') 
        
        
#Scraping Contributors Count-


    try: 
        cont_tags = driver.find_element(By.XPATH,"//*[contains(text(),'    Contributors ')]")
        Count.append(cont_tags.text.replace('Contributors',''))
    except NoSuchElementException:
        Count.append('-') 

        
#Scraping Languages-
    try:
        language = driver.find_element(By.XPATH,'//span[@class="color-fg-default text-bold mr-1"]')
        Language.append(language.text)
    except NoSuchElementException:
        Language.append('-') 
```

```python
print(len(Repository),len(Description),len(Count),len(Language),len(url))
```

```python
df=pd.DataFrame({'Repository Name':Repository,'Repository Description':Description,'Contributors Count':Count,'Language':Language})
df
```
