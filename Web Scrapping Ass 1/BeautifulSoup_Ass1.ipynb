# Question 1

Request Server to get source code of page

```python
page=requests.get('https://en.wikipedia.org/wiki/Main_Page')
```

Page Content

```python
soup=BeautifulSoup(page.content)
```

Scraping all header Tags

```python
header_tags=[]

for header in soup.find_all(["h1","h2","h3","h4","h5","h6"]):

    header_tags.append(header.name+"   "+header.text.strip())
```

# Question 2

```python
page=requests.get('https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc')
page1=requests.get('https://www.imdb.com/search/title/?groups=top_100&sort=user_rating,desc&start=51&ref_=adv_nxt')
```

```python
soup=BeautifulSoup(page.content)
soup1=BeautifulSoup(page1.content)
```

```python
Movie_Name=[]

for i in soup.find_all('h3',class_="lister-item-header"):
    Movie_Name.append(i.text.split("\n")[2])
for i in soup1.find_all('h3',class_="lister-item-header"):
    Movie_Name.append(i.text.split("\n")[2])
```

```python
Year_of_Release=[]
for i in soup.find_all('span',class_="lister-item-year text-muted unbold"):
    Year_of_Release.append(i.text)

for i in soup1.find_all('span',class_="lister-item-year text-muted unbold"):
    Year_of_Release.append(i.text)
```

```python
Rating=[]
for i in soup.find_all('div',class_="inline-block ratings-imdb-rating"):
    Rating.append(i.get('data-value'))

for i in soup1.find_all('div',class_="inline-block ratings-imdb-rating"):
    Rating.append(i.get('data-value'))
```

```python
print(len(Movie_Name),len(Rating),len(Year_of_Release))
```

```python
import pandas as pd
df=pd.DataFrame({'Movie Name':Movie_Name,'Rating':Rating,'Year of Release':Year_of_Release})
df
```

# Question 5

A) Top 10 Men ODI teams

```python
page=requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
soup=BeautifulSoup(page.content)
```

```python
Name=[]
for i in soup.find_all('span',class_="u-hide-phablet"):
    Name.append(i.text)
    
Matches=[]
a=soup.find('td',class_="rankings-block__banner--matches")
Matches.append(a.text)
for i in soup.find_all('tr',class_="table-body"):
    Matches.append(i.text.split('\n')[7])
    

Points=[]
a=soup.find('td',class_="rankings-block__banner--points")
Points.append(a.text)
for i in soup.find_all('tr',class_="table-body"):
    Points.append(i.text.split('\n')[8])

Rating=[]
a=soup.find('td',class_="rankings-block__banner--rating u-text-right")
Rating.append(a.text.replace('\n',''))
for i in soup.find_all('td',class_="table-body__cell u-text-right rating"):
    Rating.append(i.text)
```

```python
import pandas as pd
df=pd.DataFrame({'Top 10 ODI Men Team':Name[0:10],'Matches':Matches[0:10],'Points':Points[0:10],'Rating':Rating[0:10]})
df
```

```python
page=requests.get('https://www.cnbc.com/world/?region=world')
soup=BeautifulSoup(page.content)
```

(i) Headings

```python
Headings=[]
for i in soup.find_all('a',class_="LatestNews-headline"):
    Headings.append(i.text)
```

(ii)Time

```python
Time=[]
for i in soup.find_all('time',class_="LatestNews-timestamp"):
    Time.append(i.text)
```

(iii)News Link

```python
url=[]
for links in soup.find_all('a',class_="LatestNews-headline"):
    url.append(links.get('href'))
```

```python
import pandas as pd
df=pd.DataFrame({'Headline':Headings[0:10],'Time':Time[0:10],'News Link':url[0:10]})
df
```

# Question 9

```python
page=requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
soup=BeautifulSoup(page.content)
```

(i) Restaurant Name

```python
Name=[]
for i in soup.find_all('a',class_="restnt-name ellipsis"):
    Name.append(i.text)
```

(ii)Cuisine

```python
Cuisine=[]
for i in soup.find_all('span',class_="double-line-ellipsis"):
    Cuisine.append(i.text.split('|')[1])
```

(iii)Location

```python
Location=[]
for i in soup.find_all('div',class_="restnt-loc ellipsis"):
    Location.append(i.text)
```

(iv)Ratings

```python
Ratings=[]
for i in soup.find_all('div',class_="restnt-rating rating-4"):
    Ratings.append(i.text)
```

(v)Image url

```python
url=[]
for i in soup.find_all('img',class_="no-img"):
    url.append(i.get('data-src'))
```

```python
print(len(Name),len(Cuisine),len(Location),len(Ratings),len(url))
```

```python
import pandas as pd
df=pd.DataFrame({'Restaurant Name':Name,'Cuisine':Cuisine,'Loaction':Location,'Ratings':Ratings,'Image Url':url})
df
```
